#### These values appear to be vague and non understandable, so we can ignore them and remove all these rows.

```python
df_clean = df_clean.replace({' ':np.nan, '..':np.nan})
```

```python
def transform(n):
    return pd.to_numeric(n, errors='ignore')
df_clean_2 = df_clean.applymap(transform)
```

```python
df_clean_2
```

```python
chosen_columns = list(chosen_vars.values())
frame_list = []
for v in chosen_columns:
    frame = df_clean_2[df_clean_2['Series name']==v]
    frame = frame.melt(id_vars=['Country code', 'Series name'])
    frame = frame.rename(columns={'Country code':'Country', 'variable':'year', 'value':v})
    frame.drop(['Series name'], axis=1, inplace=True)
    frame_list.append(frame)
```

```python
from functools import reduce
```

```python
def join(l,r):
    return pd.merge(l, r, on=['Country', 'year'], how='outer')
df_new = reduce(join,frame_list)
```

```python
df_new
```

```python
df_new.isna().sum()
```

#### Now remove the columns having more number of NaN values

```python
for i in df_new.columns:
    if df_new[i].isna().sum()>3000:
        df_new.drop([i], axis=1, inplace=True)
```

```python
years_count_missing = dict.fromkeys(df_new['year'].unique(), 0)
for ind, row in df_new.iterrows():
    years_count_missing[row['year']] += row.isna().sum()
years_missing_sorted = dict(sorted(years_count_missing.items(), key=lambda item: item[1]))
print("Missing values by year :")
for key, val in years_missing_sorted.items():
    print(key, ":", val)
```

```python
print("Initially number of missing values in whole dataset :",df_new.isna().sum().sum())
print("Number of rows initially :",df_new.shape[0])
df_new = df_new[(df_new['year']>=1990) & (df_new['year']<=2008)]
print("Number of missing values after filtering years :",df_new.isna().sum().sum())
print("Number of rows after filtering :",df_new.shape[0])
```

```python
country_count_missing = dict.fromkeys(df_new['Country'].unique(), 0)
for ind, row in df_new.iterrows():
    country_count_missing[row['Country']] += row.isna().sum()
country_missing_sorted = dict(sorted(country_count_missing.items(), key=lambda item: item[1]))
print("Missing values by Country :")
for key, val in country_missing_sorted.items():
    print(key, ":", val)
```

```python
print("Initially number of missing values in whole dataset :",df_new.isna().sum().sum())
print("Number of rows initially :",df_new.shape[0])
filter = []
for k,v in country_missing_sorted.items():
    if v<100:
        filter.append(k)
        
df_new = df_new[df_new['Country'].isin(filter)]
print("Number of missing values after filtering countries :",df_new.isna().sum().sum())
print("Number of rows after filtering :",df_new.shape[0])
```

```python
df_new.isna().sum()
```

```python
to_drop = []
for i in df_new.columns:
    if df_new[i].isna().sum()>1300:
        to_drop.append(i)
df_new = df_new.drop(to_drop, axis=1)
print("Now missing values :")
df_new.isna().sum()
```

```python
df_new.dropna(axis=0, how='any', inplace=True)
```
